# 数据收集之爬虫

## 下载远程文件实例
`urllib.request.urlretrieve(url, filename=None, reporthook=None, data=None)`
- url:外部或者本地的url
- filename：保存到本地的文件路径，若未指定urllib会生成一个临时文件来保存数据
- reporthook:回调函数。当连接上服务器以及相应的数据块传输完毕的时候会触发该回调,可以利用该回调函数来显示下载进度。
- data:指post到服务器的数据。该方法返回一个包含两个元素的元组(filename, headers):
    - filename表示保存到本地的路径
    - header表示服务器的响应头。

```python
from urllib.request import urlretrieve

url = "https://s3.amazonaws.com/udacity-sdc/notMNIST_train.zip"
file = "notMNIST_train.zip"

def download_status(blocks, block_size, file_size):
    '''
    下载进度
        blocks     : 已下载的数据块
        block_size : 数据块的大小
        file_size  : 远程文件的大小
    '''
    percent = blocks * block_size / file_size
    print("Dowloading {:.2f}%. ".format(percent * 100))

urlretrieve(url,file, download_status)
```
